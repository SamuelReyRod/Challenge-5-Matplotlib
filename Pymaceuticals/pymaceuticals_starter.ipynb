```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
complete_data = pd.merge(study_results,mouse_metadata, on='Mouse ID',how='outer')
# Display the data table for preview
complete_data
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary DataFrame.

sum_mean_statistics = complete_data.groupby('Drug Regimen')['Tumor Volume (mm3)'].mean()
sum_median_statistics = complete_data.groupby('Drug Regimen')['Tumor Volume (mm3)'].median()
sum_variance_statistics = complete_data.groupby('Drug Regimen')['Tumor Volume (mm3)'].var()
sum_stdev_statistics = complete_data.groupby('Drug Regimen')['Tumor Volume (mm3)'].std()
sum_sem_statistics = complete_data.groupby('Drug Regimen')['Tumor Volume (mm3)'].sem()

sum_statistics = pd.DataFrame({'Mean Tumor Volume': sum_mean_statistics,'Median Tumor Volume': sum_median_statistics,
                               'Tumor Volume Variance': sum_variance_statistics,'Tumor Volume Std. Dev.': sum_stdev_statistics,
                               'Tumor Volume Std. Err.': sum_sem_statistics})

sum_statistics
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, 
# and SEM of the tumor volume for each regimen
# Using the aggregation method, produce the same summary statistics in a single line.

summary_statistics = complete_data.groupby('Drug Regimen')['Tumor Volume (mm3)'].agg(['mean', 'median', 'var', 'std','sem'])
summary_statistics = summary_statistics.rename(columns={'mean': 'Mean Tumor Volume',
                                                       'median':'Median Tumor Volume',
                                                       'var':'Tumor Volume Variance',
                                                       'std':'Tumor Volume Std. Dev.',
                                                       'sem':'Tumor Volume Std. Err.'})
summary_statistics
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse

#Filtering the 4 regimens
treatment_regimens = ['Capomulin', 'Ramicane', 'Infubinol','Ceftamin']
selected_treatment = complete_data.loc[complete_data['Drug Regimen'].isin(treatment_regimens)]
#sorting on timepoint descending...this will yield the last instance of timepoints and then removing all duplicates
max_timepoint = selected_treatment.sort_values(['Timepoint'], ascending=False)
max_timepoint = max_timepoint.drop_duplicates(subset=['Mouse ID'])
last_tumor_volumes = max_timepoint[['Mouse ID', 'Tumor Volume (mm3)']]
last_tumor_volumes_list = last_tumor_volumes.to_records(index=False).tolist()
last_tumor_volumes_df = pd.DataFrame(last_tumor_volumes_list, columns=['Mouse ID', 'Last Tumor Volume'])            
last_tumor_volumes_df

# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
new_clean_merge = pd.merge(complete_data,last_tumor_volumes_df,on='Mouse ID')
new_clean_merge.head()
```
